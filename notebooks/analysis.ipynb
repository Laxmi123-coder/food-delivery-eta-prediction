```python
df['Time_taken'] = df['Time_taken(min)'].str.extract(r'(\d+)').astype(float)
print(df['Time_taken'].head(10))
```

```python
df = df.dropna()
print("Shape after cleaning:", df.shape)
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return R * 2 * atan2(sqrt(a), sqrt(1-a))

df['distance_km'] = df.apply(
    lambda r: haversine(
        r['Restaurant_latitude'], r['Restaurant_longitude'],
        r['Delivery_location_latitude'], r['Delivery_location_longitude']
    ), axis=1
)

print(df['distance_km'].describe())
```

```python
print("Missing values:", df.isnull().sum().sum())
print("Total rows:", len(df))
print("distance_km sample:", df['distance_km'].head())
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Time_taken'], bins=30, color='#ff6b2b')
plt.title('Distribution of Delivery Time')
plt.xlabel('Time (minutes)')
plt.savefig('../images/chart1_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Road_traffic_density', y='Time_taken', data=df)
plt.title('Traffic Impact on Delivery Time')
plt.savefig('../images/chart2_traffic.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Weatherconditions', y='Time_taken', data=df)
plt.xticks(rotation=45)
plt.title('Weather Impact on Delivery Time')
plt.savefig('../images/chart3_weather.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
df = df.dropna(subset=['Road_traffic_density'])
print("Rows after fix:", len(df))
print("Traffic values:", df['Road_traffic_density'].unique())
```
